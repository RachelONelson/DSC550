# Scenario:
I am a pokemon master and want to figure out how many unique pokemon belong to each type. I want to display this
info in a chart.

```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Import data from csv
df=pd.read_csv("D:\machine-learning-with-python-cookbook\Pokemon.csv")
df_type = df['Type 1'].value_counts()
df_type = pd.DataFrame(df_type)
df_type.head(5)

# Im using this data to create simple x and y
```

```python
# Bar Chart Example
# Create Chart, increased width of chart a bit
ax = df_type.plot.barh(width=.7,figsize=(10, 6))
# Add Labels
plt.xlabel("Type")
plt.ylabel("# of Pokemon")
plt.title("# Pokemon by Type",fontsize=14)
plt.legend()
# added data labels for each value
for index, value in enumerate(df_type['Type 1']):
    plt.text(value, index, str(value))
plt.text(-0.6,-5,'Source: https://www.kaggle.com/abcsds/pokemon')
plt.show()
```

```python
# Line Chart Example
# Create Chart - default chart is fine
ax = df_type.plot.line(lw=2.5,color='blue',marker='o',figsize=(10, 6))
plt.ylim(0,120)
# Add Labels
plt.xlabel("Pokemon Type")
plt.ylabel("# of Pokemon")
plt.title("# Pokemon by Type",fontsize=14)
#Background color
ax.yaxis.grid(color='orange')
#put the text on bottom of chart because it fits in whitespace
plt.text(-0.6,1,'Source: https://www.kaggle.com/abcsds/pokemon')
plt.show()
```

```python
x_pos = [i for i, _ in enumerate(df_type)]

# Pie Chart Example
# Create Chart
ax = df_type.plot.pie(subplots=True, figsize=(10, 6))
# Add Labels
plt.xlabel("Pokemon Type")
plt.ylabel("# of Pokemon")
plt.title("# Pokemon by Type",fontsize=14)
# Moving legend so it does not collide with the chart
plt.legend(loc="lower center", bbox_to_anchor=(0.5, -0.50), ncol= 2)
plt.text(-0.6,-2.6,'Source: https://www.kaggle.com/abcsds/pokemon')
plt.show()
```
